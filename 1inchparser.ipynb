```python
"""singular value decomposition of a matrix"""
import numpy as np
def svd(A):
    """Singular Value Decomposition

    Factorizes the matrix A into U*S*V such that U*S*V = A, where S is a diagonal matrix of singular values and V is a column matrix of eigenvectors.
    """
    U, s, V = np.linalg.svd(A)
    return U, s, V





```
